```python
import csv

txtfile = open('/home/jink/caffe/data/fashionAILabel/txtfile.txt','w')

with open('/home/jink/caffe/data/fashionAILabel/label.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        labelNum = row[2].index('y')
        txtfile.write(row[0] + ' ' +str(labelNum)+'\n')
    f.close()
txtfile.close()




import os
road = '/home/jink/caffe/data/fashionAILabel/'
Classes = [('coat_length_labels',8),
           ('collar_design_labels',5),
           ('lapel_design_labels',5),
           ('neck_design_labels',5),
           ('neckline_design_labels',10),
           ('pant_length_labels',6),
           ('skirt_length_labels',6),
           ('sleeve_length_labels',9)
           ]

for (item, Classnum) in Classes:
        #Classnum = label
    folder = road+item
    print folder
    if not os.path.exists(folder):
        os.mkdir(folder)
            
    img = {}
    classcount = {}
    for i in range(Classnum):
        img[str(i)]=[]
        classcount[str(i)] = 0
    txtfile = open(road+ 'txtfile.txt','r')
    while True:
        lines = txtfile.readline()
        if not lines:
            break
        if (lines.split()[0].split('/')[-2] == item):
            img[lines.split()[1]].append(lines)
      
    valF = folder + '/val.txt'
    trainF = folder + '/train.txt'
    for j in range(5):
        ftrain = open(trainF.replace('train', 'trainfold'+str(j+1)), 'w')
        fval = open(valF.replace('val', 'valfold'+str(j+1)), 'w')
        for i in range(Classnum):
            classcount[str(i)] = len(img[str(i)])
            val = img[str(i)][classcount[str(i)]*j/5:classcount[str(i)]*(j+1)/5] 
            train = img[str(i)][:classcount[str(i)]*j/5] + img[str(i)][classcount[str(i)]*(j+1)/5:]
            ftrain.write(''.join(train))
            fval.write(''.join(val))
        ftrain.close()
        fval.close()
    txtfile.close()
                

# 将多个trainflod1合并成一个txt
# design类合成一个，length类合成一个
```
